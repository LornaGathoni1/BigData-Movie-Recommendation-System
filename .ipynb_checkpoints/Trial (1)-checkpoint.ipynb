## **1.2 Ingesting Data into Kafka**

In this sub-step, we send the MovieLens data to a Kafka topic (`tmdb_topic`) using a Kafka for real-time processing.
The steps involved in this process are as follows:

```python
from confluent_kafka import Producer
import json

# Kafka configuration
kafka_server = "localhost:9092"  # Pointing to local Kafka instance
topic = "tmdb_topic"

# Kafka Producer configuration
conf = {
    'bootstrap.servers': kafka_server,
    'client.id': 'movielens-producer'
}

# Create Producer instance
producer = Producer(conf)

# Function to send data to Kafka topic
def produce_data():
    # Example data - sending JSON movie data to Kafka
    for i in range(10):
        movie_data = json.dumps({"movie_id": i, "title": f"Movie {i}"})
        producer.produce(topic, key="key", value=movie_data)
        producer.flush()  # Ensure the data is sent
        print(f"Message {i} sent to Kafka topic {topic}") # Print message after sending data

produce_data()  # Call the function to send data to Kafka
```

Visualization:

Visualization refers to the graphical representation of data, information, or concepts in a visual format, such as charts, graphs, maps, or infographics. The goal of visualization is to make complex data more accessible, understandable, and usable by presenting it in a way that is easy to interpret. It allows individuals to quickly grasp patterns, trends, and insights that might be harder to identify in raw data.



1. Top 10 Genres

Horizontal Bar Chart: This visualization displays the top 10 most frequent genres in the dataset. It helps to quickly identify which genres are dominant by showing the number of movies associated with each genre in a clear and readable format.

```python
import matplotlib.pyplot as plt

# Assuming 'genres' column is a list of genres for each movie
genres_data = movies.explode('genres')  # Split genres into separate rows
top_genres = genres_data['genres'].value_counts().head(15)

plt.figure(figsize=(10, 6))
top_genres.plot(kind='barh', color='skyblue')
plt.title("Top 10 Genres")
plt.xlabel("Number of Movies")
plt.ylabel("Genres")
plt.gca().invert_yaxis()  # Flip order for better readability
plt.show()
```

2. Top 15 Movies by ID (Mock Popularity)

Vertical Bar Chart: This visualization shows the top 15 movies sorted by their ID values, using the id column as a mock measure of popularity. It provides a quick view of movies with the highest IDs, represented in a vertical bar chart for easy comparison.

```python
import matplotlib.pyplot as plt

# Sort by 'id' (acting as a mock metric for popularity)
top_movies = movies[['title', 'id']].sort_values(by='id', ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(top_movies['title'], top_movies['id'], color='skyblue')
plt.title("Top 15 Movies by ID (Mock Popularity)")
plt.xlabel("Movies")
plt.ylabel("ID (as Mock Popularity)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

5. Top 15 Movies by Tags Length

Horizontal Bar Chart: This visualization displays the top 15 movies with the longest tags (based on the character count of the tags column). It provides insights into which movies have the most detailed or descriptive tags, helping to understand tagging patterns in the dataset.

```python
# Add a column for tags length
movies['tags_length'] = movies['tags'].apply(lambda x: len(x) if isinstance(x, str) else 0)

# Sort by 'tags_length' and get the top 15
top_tags = movies[['title', 'tags_length']].sort_values(by='tags_length', ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.barh(top_tags['title'], top_tags['tags_length'], color='teal')
plt.title("Top 15 Movies by Tags Length")
plt.xlabel("Tags Length (Characters)")
plt.ylabel("Movies")
plt.gca().invert_yaxis()  # Flip for readability
plt.show()
```

6. Most Common Words in Tags

Word Cloud: This visualization generates a word cloud of the most frequently used words in the tags column. It provides a quick and visually appealing way to identify prominent themes or keywords associated with movies in the dataset.

```python
from wordcloud import WordCloud

# Combine all tags into a single string
all_tags = " ".join(movies['tags'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_tags)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Tags")
plt.show()
```
